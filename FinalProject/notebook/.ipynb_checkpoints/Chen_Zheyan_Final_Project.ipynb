B) ***Data Exploration and Pre-processing***

```python
from sklearn.preprocessing import OneHotEncoder
import os
import random
import glob
import math

onehot_encoder = OneHotEncoder(sparse=False)
enc_y = onehot_encoder.fit_transform(y.values)

dir_path = '../Data/images'
class_img_set = []

# get # of images in each class folder
for path in os.listdir(dir_path):
    if path != '.DS_Store':
        sub_path = os.path.join(dir_path, path)
        count = 0
        for p in os.listdir(sub_path):
            if os.path.isfile(os.path.join(sub_path, p)):
                count +=1 
        class_img_set.append((sub_path, count))
# print(class_img_set)
tr_imgs = []
val_imgs = []
ts_imgs = []

for class_img_t in class_img_set:
    all_imgs = glob.glob(class_img_t[0]+"/*.jpg")
    # take ceiling of 70% of all images
    tr_lt = random.sample(all_imgs, math.ceil(0.7*class_img_t[1]))
    tr_imgs.extend(tr_lt)
    # get the rest of images
    remain_imgs = [img for img in all_imgs if img not in tr_lt]
    # take ceiling of 15% of remain images 
    val_lt = random.sample(remain_imgs, math.ceil(0.15*class_img_t[1]))
    val_imgs.extend(val_lt)
    # take the rest as test images
    ts_imgs.extend([img for img in remain_imgs if img not in val_lt])
    
print(f"Train set size is "+ str(len(tr_imgs)))
print(f"Val set size is "+ str(len(val_imgs)))
print(f"Test set size is "+ str(len(ts_imgs)))

```

```python
from PIL import Image
import cv2
import shutil

# Remove 3 folders
if os.path.isdir("../Data/train"):
    shutil.rmtree("../Data/train")
if os.path.isdir("../Data/validation"):
    shutil.rmtree("../Data/validation")
if os.path.isdir("../Data/test"):
    shutil.rmtree("../Data/test")
    
# create 3 different folders if not existed 
if not os.path.isdir("../Data/train"):
    os.makedirs("../Data/train")
if not os.path.isdir("../Data/validation"):
    os.makedirs("../Data/validation")
if not os.path.isdir("../Data/test"):
    os.makedirs("../Data/test")
    
# resize train set and write back to train folder
for img in tr_imgs:
    image = cv2.imread(img)
    resized = cv2.resize(image, (224, 224))
    filenames = img.split('/')
    if not os.path.isdir("../Data/train/"+(filenames[-2])):
        os.makedirs("../Data/train/"+filenames[-2])
    cv2.imwrite("../Data/train/"+filenames[-2]+"/"+"224_"+filenames[-1], resized)
    
# resize validation set and write back to train folder
for img in val_imgs:
    image = cv2.imread(img)
    resized = cv2.resize(image, (224, 224))
    filenames = img.split('/')
    if not os.path.isdir("../Data/validation/"+(filenames[-2])):
        os.makedirs("../Data/validation/"+filenames[-2])
    cv2.imwrite("../Data/validation/"+filenames[-2]+"/"+"224_"+filenames[-1], resized)
    
# resize test set and write back to train folder
for img in ts_imgs:
    image = cv2.imread(img)
    resized = cv2.resize(image, (224, 224))
    filenames = img.split('/')
    if not os.path.isdir("../Data/test/"+(filenames[-2])):
        os.makedirs("../Data/test/"+filenames[-2])
    cv2.imwrite("../Data/test/"+filenames[-2]+"/"+"224_"+filenames[-1], resized)

```

C) ***Transfer Learning***

***EfficientNetB0 with crop; zoom; rotation; translation; flip; contrast***

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np

training_set = image_dataset_from_directory('../Data/train', image_size = (224,224), labels = 'inferred',label_mode = 'categorical')
val_set = image_dataset_from_directory('../Data/validation', image_size = (224,224), labels = 'inferred',label_mode = 'categorical')

# Empirical Regularizaton using keras.layers API Image augmentation
img_augmentation = Sequential(
    [
        layers.Cropping2D((33,33)),
        layers.RandomZoom(.5,.2),
        layers.RandomRotation(factor=0.15),
        layers.RandomTranslation(height_factor=0.1, width_factor=0.1),
        layers.RandomFlip(),
        layers.RandomContrast(factor=0.1),
    ],
    name="img_augmentation",
)

# Plot Training and Val error VS epochs
def plot_hist(hist):
    plt.plot(hist.history["loss"])
    plt.plot(hist.history["val_loss"])
    plt.title("model error")
    plt.ylabel("error")
    plt.xlabel("epoch")
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()

def build_b0_model(num_classes):
    inputs = layers.Input(shape=(224,224,3))
    x = img_augmentation(inputs)
    model = EfficientNetB0(include_top=False, input_tensor=x, weights="imagenet")

    # Freeze the pretrained weights
    model.trainable = False

    # Rebuild top / last fully connected layer
    x = layers.GlobalAveragePooling2D(name="avg_pool")(model.output)
    x = layers.BatchNormalization()(x)

    top_dropout_rate = 0.2
    x = layers.Dropout(top_dropout_rate, name="top_dropout")(x)
    
    # last layer with 2 activation functions
    layerA_out = layers.Dense(num_classes, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(layerA_out)

    # Compile
    model = tf.keras.Model(inputs, outputs, name="EfficientNet")
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-2)
    model_metrics = ['accuracy',tf.keras.metrics.Precision(),tf.keras.metrics.Recall()]
    model.compile(
        optimizer=optimizer, loss="categorical_crossentropy", metrics=model_metrics
    )
    return model


b0_model = build_b0_model(20)

# Early Stopping based on validation
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=10,min_delta = 0.01,mode ="min",verbose=1)

b0_hist = b0_model.fit(training_set, epochs=70, validation_data=val_set, verbose=2, batch_size =5, callbacks = [early_stopping])
plot_hist(b0_hist)

```

```python
from sklearn.metrics import classification_report

test_set = image_dataset_from_directory('../Data/test', image_size = (224,224), labels = 'inferred',label_mode = 'categorical')
y_prob = b0_model.predict(test_set, verbose=1)
y_pred = np.argmax(y_prob, axis =1).tolist()

# get one-hot encoded label from tensorflow dataset
# convert label from numpy to list
one_hot_enc_label = np.concatenate([y for x, y in test_set], axis=0).tolist()
y_true = []
for row in range(len(one_hot_enc_label)):
    for idx in range(len(one_hot_enc_label[row])):
        if one_hot_enc_label[row][idx] == 1.0:
            y_true.append(idx)
            break
print(classification_report(y_true, y_pred))
```

C) ***Transfer Learning***

***VGG16 with random rotation, random flip, random contrast, random translation and batch == 5***

```python
from tensorflow.keras.applications.vgg16 import VGG16

# pre-trained VGG16 without last layer
def build_vgg16_model(num_classes):
    inputs = layers.Input(shape=(224,224,3))
    x = img_augmentation(inputs)
    
    # Must preprocess input for VGG16
    x = tf.keras.applications.vgg16.preprocess_input(x)
    model = VGG16(include_top=False, input_tensor=x, weights="imagenet")

    # Freeze the pretrained weights
    model.trainable = False

    # Rebuild top / last fully connected layer
    x = layers.GlobalAveragePooling2D(name="avg_pool")(model.output)
    x = layers.BatchNormalization()(x)

    top_dropout_rate = 0.2
    x = layers.Dropout(top_dropout_rate, name="top_dropout")(x)
    
    # last layer with 2 activation functions
    layerA_out = layers.Dense(num_classes, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(layerA_out)

    # Compile
    model = tf.keras.Model(inputs, outputs, name="vgg16")
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-2)
    model_metrics = ['accuracy',tf.keras.metrics.Precision(),tf.keras.metrics.Recall()]
    model.compile(
        optimizer=optimizer, loss="categorical_crossentropy", metrics=model_metrics
    )
    return model


vgg16_model = build_vgg16_model(20)

# Early Stopping based on validation
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=10,min_delta = 0.01,mode ="min",verbose=1)

vgg16_hist = vgg16_model.fit(training_set, epochs=50, validation_data=val_set, verbose=2, batch_size =5, callbacks = [early_stopping])
plot_hist(vgg16_hist)
```

```python
test_set = image_dataset_from_directory('../Data/test', image_size = (224,224), labels = 'inferred',label_mode = 'categorical')
y_prob = vgg16_model.predict(test_set, verbose=1)
y_pred = np.argmax(y_prob, axis =1).tolist()

# get one-hot encoded label from tensorflow dataset
# convert label from numpy to list
one_hot_enc_label = np.concatenate([y for x, y in test_set], axis=0).tolist()
y_true = []
for row in range(len(one_hot_enc_label)):
    for idx in range(len(one_hot_enc_label[row])):
        if one_hot_enc_label[row][idx] == 1.0:
            y_true.append(idx)
            break
print(classification_report(y_true, y_pred))
```
